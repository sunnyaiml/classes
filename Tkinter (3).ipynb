```python
from tkinter import *   
  
top = Tk()  
  
top.geometry("200x200")  
  
checkvar1 = IntVar()  
  
checkvar2 = IntVar()  
  
checkvar3 = IntVar()  
  
def fun(i, j):  
    select = []
    text = j.cget("text")
    if i.get() == 1:
        select.append(text)
        messagebox.OK
    else:
        messagebox.showinfo("Alart",f'Unselected Options {text}')
        
    
def sumited():
    checkvar1.set(0)
    checkvar2.set(0)
    checkvar3.set(0)
    messagebox.showinfo("Success", "All selections have been Submited")
    
def check():
    checkvar1.set(0)
    checkvar2.set(0)
    checkvar3.set(0)
    messagebox.showinfo("Cleared", "All selections have been cleared")
    
  
chkbtn1 = Checkbutton(top, text = "C",command=lambda: fun(checkvar1, chkbtn1), variable = checkvar1, onvalue = 1, offvalue = 0, height = 2, width = 10)  
  
chkbtn2 = Checkbutton(top, text = "C++",command=lambda: fun(checkvar2, chkbtn2), variable = checkvar2, onvalue = 1, offvalue = 0, height = 2, width = 10)  
  
chkbtn3 = Checkbutton(top, text = "Java",command=lambda: fun(checkvar3, chkbtn3), variable = checkvar3, onvalue = 1, offvalue = 0, height = 2, width = 10)  

clear = Button(top, text = "check_element",command=lambda:check(), activeforeground = "yellow",activebackground = "pink",pady = 10)

submiting = Button(top, text = "Submit",command=lambda: sumited(), activeforeground = "yellow",activebackground = "blue",pady = 10)

chkbtn1.pack()  
  
chkbtn2.pack()  
  
chkbtn3.pack() 

clear.pack(side = BOTTOM) 
submiting.pack()
  
top.mainloop()  
```

```python
#Example: A simple calculator
import tkinter as tk  
from functools import partial  
   
   
def call_result(label_result, n1, n2):  
    num1 = (n1.get())  
    num2 = (n2.get())  
    result = int(num1)+int(num2)  
    label_result.config(text="Result = %d" % result)  
    return  
   
root = tk.Tk()  
root.geometry('400x200+100+200')  
  
root.title('Calculator')  
   
number1 = tk.StringVar()  
number2 = tk.StringVar()  
  
labelNum1 = tk.Label(root, text="A").grid(row=1, column=0)  
  
labelNum2 = tk.Label(root, text="B").grid(row=2, column=0)  
  
labelResult = tk.Label(root)  
  
labelResult.grid(row=7, column=2)  
  
entryNum1 = tk.Entry(root, textvariable=number1).grid(row=1, column=2)  
  
entryNum2 = tk.Entry(root, textvariable=number2).grid(row=2, column=2)  
  
call_result = partial(call_result, labelResult, number1, number2)  
  
buttonCal = tk.Button(root, text="Calculate", command=call_result).grid(row=3, column=0)  
  
root.mainloop()  
```

```python
from tkinter import *
from tkinter import messagebox, simpledialog 

top = Tk()
top.geometry("300x300")

lbl = Label(top, text="A list of favourite countries...")
lbl.pack()


listbox = Listbox(top)
listbox.insert(1, "India")
listbox.insert(2, "USA")
listbox.insert(3, "Japan")
listbox.insert(4, "Australia")
listbox.pack()


def add():
    while True:
        coun = simpledialog.askstring("Input", "Enter the country name:")
        country = str(coun).strip
        if len(country)<0:
            response = messagebox.askyesno("ENTER VALUE SHOULD NOT BE EMPTY")
            country = simpledialog.askstring("Input", "Enter the country name:")
            if country:
                listbox.insert(END, country)
            else:
                break 
            response = messagebox.askyesno("Add More", "Do you want to add more countries?")
            if not response:
                break


def delete_item():
    selected = listbox.curselection()
    if selected:
        listbox.delete(selected)
    else:
        messagebox.showinfo("No selection", "Please select an item to delete.")

# Buttons
btn_add = Button(top, text="Add", command=add)
btn_delete = Button(top, text="Delete", command=delete_item)

btn_add.pack(pady=5)
btn_delete.pack(pady=5)

top.mainloop()
```
